```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stock = 'AAPL'
data = yf.download(stock, start="2023-01-01", end="2024-07-01")
print(data.head())
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Close'], label="AAPL closing values")
plt.title("Apple stocks")
plt.xlabel("Date")
plt.ylabel("Price(USD)")
plt.legend()
plt.grid()
plt.show()
```

```python
data.to_csv('apple_data_stoc_prizes')
```

# section 4 fetch live data

```python
import time
ticker = yf.Ticker("AAPL")

while True:
    data = ticker.history(period = "1d", interval="1m")
    current_price = data['Close'].iloc[-1]
    current_time = data.index[-1]
    print(f"{current_time} Current AAPL prize: ${current_price}")
    time.sleep(60)
    
```
