```python
# Import the libraries
import requests                  # For making HTTP requests
from bs4 import BeautifulSoup     # For parsing HTML content
import pandas as pd               # For creating, storing and manipulating DataFrames
```

### 2. Send a request to IMDb

```python
url="https://www.goodreads.com/quotes"
```

```python
response=requests.get(url)
response
```

```python
headers={    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
```

```python
response=requests.get(url,headers=headers)
response
```

```python
if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
soup=BeautifulSoup(response.content,"html.parser")
```

```python
soup
```

```python
print(soup.prettify())
```

```python
soup.a
```

```python
soup.find_all('a')
```

```python
quotes=soup.find_all("div" ,class_='quote')
```

```python
quotes
```

```python
 print(f"Number of qoutes found: {len(quotes)}")
```

```python
quotes
```

```python
quotes_data=[]
```

```python
for quote in quotes:
    
    quote_text_div = quote.find('div', class_='quoteText').get_text(strip=True)
    author = quote.find('span', class_='authorOrTitle').get_text(strip=True)

    
    quotes_data.append({
        'Quote': quote_text_div,
        'Author': author,
        
    })
```

```python
print(quotes_data)
```

```python
df=pd.DataFrame(quotes_data)
```
